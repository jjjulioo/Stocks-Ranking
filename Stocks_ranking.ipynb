```python
def calculate_financial_metrics(ticker_symbol):
    # Initialize the Ticker object
    ticker = yf.Ticker(ticker_symbol)

    try:
        # Income Statement
        income_stmt = ticker.income_stmt
    except Exception:
        income_stmt = None

    try:
        # Sales
        sales = income_stmt.iat[-1, 0] if income_stmt is not None else 0
    except Exception:
        sales = 0

    try:
        # Net income
        net_income = income_stmt.iat[21, 0] if income_stmt is not None else 0
    except Exception:
        net_income = 0

    try:
        # % Sales
        percentage_sales = (net_income / sales) * 100 if sales != 0 else 0
    except Exception:
        percentage_sales = 0

    try:
        # Market Cap
        market_cap = ticker.info.get('marketCap', 0)
    except Exception:
        market_cap = 0

    try:
        # P/E
        pe_ratio = ticker.info.get('trailingPE', 0)
    except Exception:
        pe_ratio = 0

    try:
        # P/Book
        price_to_book = ticker.info.get('priceToBook', 0)
    except Exception:
        price_to_book = 0

    try:
        # P/Sales
        price_to_sales = ticker.info.get('priceToSalesTrailing12Months', 0)
    except Exception:
        price_to_sales = 0

    try:
        # Cash
        cash = ticker.info.get('totalCash', 0)
    except Exception:
        cash = 0

    try:
        # Debt
        total_debt = ticker.info.get('totalDebt', 0)
    except Exception:
        total_debt = 0

    try:
        # Net Cash
        net_cash = cash - total_debt if cash != 0 and total_debt != 0 else 0
    except Exception:
        net_cash = 0

    try:
        # Cash yield
        cash_yield = net_cash / market_cap if market_cap != 0 else 0
    except Exception:
        cash_yield = 0

    try:
        # CAGR Sales
        end_revenue = sales
        start_revenue = income_stmt.iat[-1, 3] if income_stmt is not None else 0
        cagr_sales = end_revenue/start_revenue if start_revenue != 0 else 0
    except Exception:
        cagr_sales = 0

    try:
        # ROE
        roe = ticker.info.get('returnOnEquity', 0)
    except Exception:
        roe = 0

    try:
        # Book tangible
        book_tangible = ticker.info.get('bookValue', 0)
    except Exception:
        book_tangible = 0

    try:
      # 4 year shares
      shares_past = income_stmt.iat[16,3] #if shares_past is not None else 0
      shares_current = income_stmt.iat[16,0] #if income_stmt is not None else 0
      cagr_shares = shares_current/shares_past*100 #if shares_past != 0 else 0
    except Exception:
      cagr_shares = 0

    try:
      # Net income CAGR 4 years
      start_income = income_stmt.iat[21, 3] if income_stmt is not None else 0
      cagr_net_income = net_income/start_income if start_income != 0 else 0
    except Exception:
      cagr_net_income = 0

    # Create dictionary with calculated metrics
    data = {
        'Ticker': ticker_symbol,
        'Sales': sales,
        'Net Income': net_income,
        '% Sales': percentage_sales,
        'Market Cap': market_cap,
        'P/E': pe_ratio,
        'P/Book': price_to_book,
        'P/Sales': price_to_sales,
        'Cash': cash,
        'Debt': total_debt,
        'Net Cash': net_cash,
        'Cash Yield': cash_yield,
        'CAGR Sales': cagr_sales,
        'ROE': roe,
        'Book Tangible': book_tangible,
        'CAGR Net Income (4 years)': cagr_net_income,
        'CAGR Shares (4 years)': cagr_shares
    }


    return data
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))

symbols = [company["Symbol"] for company in df]


tickers = []

for i in range(len(symbols[0])):
  if type(symbols[0][i]) == str:
    tickers.append(symbols[0][i])
```

```python
data = []

for ticker in tickers:
  metrics = calculate_financial_metrics(ticker)
  data.append(metrics)

def create_financial_metrics_dataframe(data, tickers):
  print(len(tickers))
  # Create DataFrame from data dictionary
  df = pd.DataFrame(data)
  return df

df = create_financial_metrics_dataframe(data, tickers)
df.fillna(0, inplace=True)
```

```python
def rank_metrics(df):
    # Ensure the DataFrame contains the 'Ticker' column
    if 'Ticker' not in df.columns:
        raise ValueError("The DataFrame must contain a 'Ticker' column.")

    # Select only the specified columns for ranking
    columns_to_rank = ['% Sales', 'P/E', 'Cash Yield', 'P/Book', 'CAGR Sales', 'ROE', 'CAGR Net Income (4 years)']

    # Convert specified columns to numeric, coercing errors to NaN
    df[columns_to_rank] = df[columns_to_rank].apply(pd.to_numeric, errors='coerce')

    # Replace NaN values with 0
    df[columns_to_rank] = df[columns_to_rank].fillna(0)

    # Initialize a DataFrame to hold the rankings
    ranked_df = pd.DataFrame(index=df.index)

    # Rank P/E and P/B in descending order (lower value gets higher rank)
    ranked_df['Rank P/E'] = df['P/E'].rank(method='min', ascending=False)
    ranked_df['Rank P/Book'] = df['P/Book'].rank(method='min', ascending=False)

    # Rank the other columns in ascending order (higher value gets higher rank)
    for col in ['% Sales', 'Cash Yield', 'CAGR Sales', 'ROE', 'CAGR Net Income (4 years)']:
        ranked_df[f'Rank {col}'] = df[col].rank(method='min', ascending=True)

    # Set rank to 1 for any zero values
    for col in columns_to_rank:
        ranked_df[f'Rank {col}'][df[col] == 0] = 1

    # Calculate the total rank by summing the ranks
    ranked_df['Total Rank'] = ranked_df.sum(axis=1)

    # Add the Ticker column to the ranked DataFrame
    ranked_df['Ticker'] = df['Ticker']

    # Reorder the columns to have Ticker first
    final_columns = ['Ticker'] + [f'Rank {col}' for col in columns_to_rank] + ['Total Rank']
    ranked_df = ranked_df[final_columns]

    return ranked_df

ranked_df = rank_metrics(df)
ranked_df
```

```python
ranked_df_ordered = ranked_df.sort_values(by='Total Rank', ascending=False)
ranked_df_ordered
```
